```python
def play_bingo(val,cards):

    val_found = False

    # check each card
    for card_id in cards.keys():

        # check columns
        for column_id in cards[card_id]['columns'].keys():

            if val in cards[card_id]['columns'][column_id].keys():

                cards[card_id]['columns'][column_id][val] = 1
                val_found = True

        # check row
        if val_found:
            for row_id in cards[card_id]['rows'].keys():

                if val in cards[card_id]['rows'][row_id].keys():

                    cards[card_id]['rows'][row_id][val] = 1
                    val_found = False

        # check for completed column
        for column_id in cards[card_id]['columns'].keys():
            
            col_sum = 0

            for val_idx in cards[card_id]['columns'][column_id]:

                col_sum = col_sum+cards[card_id]['columns'][column_id][val_idx]

            if col_sum == 5:
                
                return True,card_id,'c',column_id,val

        # check for completed row
        for row_id in cards[card_id]['rows'].keys():
            
            r_sum = 0

            for val_idx in cards[card_id]['rows'][row_id]:

                r_sum = r_sum+cards[card_id]['rows'][row_id][val_idx]

            if r_sum == 5:
                
                return True,card_id,'r',row_id,val
            
    return False,0,0,0,0
```
